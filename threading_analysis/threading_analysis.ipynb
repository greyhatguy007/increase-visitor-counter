```python
import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt

y1 = [58.583, 66.153, 79.749, 67.131, 64.029, 66.813, 67.027]
x1 = [0,1,2,3,4,5,6]
z1 = [30,30,30,30,30,30,30]

y2 = [44.102, 55.461, 33.596, 39.266, 33.573, 39.806]
x2 = [0,1,2,3,4,5]
z2 = [15,15,15,15,15,15]

y3 = [25.625, 27.517, 29.561, 28.604,31.248, 29.659]
x3 = [0,1,2,3,4,5]
z3 = [10,10,10,10,10,10]
ax = plt.axes(projection='3d')
ax.set_xlabel('count')
ax.set_ylabel('threads')
ax.set_zlabel('time')
ax.plot(z1, x1, y1, label = "count : 30 ", color='red')
ax.plot(z2, x2, y2, label = "count : 15 ", color='blue')
ax.plot(z3, x3, y3, label = "count : 10 ", color='green')

ax.view_init(0, 290)
plt.show()
```
